```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
# Dönüşümler (resimleri tensörlere çevirip normalize etmek için)
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # Normalizasyon
])

# Eğitim ve test veri setlerini indirme
train_dataset = datasets.GTSRB(root='./data', split='train', download=True, transform=transform)
test_dataset = datasets.GTSRB(root='./data', split='test', download=True, transform=transform)

# DataLoader ile veri setlerini yükleme
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
# Örnek veri ayırma işlemi (görüntü ve etiketler için)
images, labels = [], []

for data, label in train_loader:
    images.append(data)
    labels.append(label)

# Veriyi NumPy array'lere çevirme
images = np.vstack([img.numpy() for img in images])
labels = np.hstack([lbl.numpy() for lbl in labels])

# Eğitim ve test olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()  # veya MinMaxScaler()
# Verilerin yeniden şekillendirilmesi
X_train_scaled = scaler.fit_transform(X_train.reshape(X_train.shape[0], -1))
X_test_scaled = scaler.transform(X_test.reshape(X_test.shape[0], -1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten

# MLP Modeli
model = Sequential([
    Dense(128, input_dim=X_train_scaled.shape[1], activation='relu'),
    Dense(64, activation='relu'),
    Dense(43, activation='softmax')  # 43 sınıf
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_scaled, y_train, epochs=10, batch_size=64, validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred_classes = np.argmax(y_pred, axis=1)

print("Accuracy:", accuracy_score(y_test, y_pred_classes))
print("Classification Report:\n", classification_report(y_test, y_pred_classes))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV

# MLP modelini oluştur ve eğit
mlp = MLPClassifier(max_iter=100)

# Hiperparametre grid araması için parametreler
param_grid = {
    'hidden_layer_sizes': [(50,), (100,)],  # Farklı katman ve nöron kombinasyonları
    'activation': ['relu', 'tanh',],                     # Aktivasyon fonksiyonları
    'learning_rate': ['constant'],                      # Öğrenme hızı
}

# GridSearchCV ile en iyi parametreleri bulma
grid_search = GridSearchCV(mlp, param_grid, cv=3, scoring='accuracy', verbose=2, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# En iyi model
best_mlp = grid_search.best_estimator_

# Model performansı
y_pred = best_mlp.predict(X_test_scaled)
print("MLP Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np
from sklearn.metrics import accuracy_score, classification_report

# RBF Katmanı tanımı
class RBFLayer(tf.keras.layers.Layer):
    def __init__(self, num_centers, gamma):
        super(RBFLayer, self).__init__()
        self.num_centers = num_centers
        self.gamma = gamma

    def build(self, input_shape):
        self.centers = self.add_weight(name='centers',
                                       shape=(self.num_centers, input_shape[-1]),
                                       initializer='random_normal',
                                       trainable=True)
        self.betas = self.add_weight(name='betas',
                                     shape=(self.num_centers,),
                                     initializer='ones',
                                     trainable=True)

    def call(self, inputs):
        diff = tf.expand_dims(inputs, axis=1) - self.centers
        l2 = tf.reduce_sum(tf.square(diff), axis=-1)
        return tf.exp(-self.gamma * l2)

# Hiperparametre optimizasyonu için parametreler
param_grid = {
    'num_centers': [5, 10, 20],  # RBF merkez sayısı
    'gamma': [0.1, 0.5, 1],     # Gamma değeri
    'batch_size': [32, 64],     # Batch boyutu
    'epochs': [10, 20],         # Epoch sayısı
}

# Hiperparametre optimizasyon döngüsü
best_accuracy = 0
best_params = None
best_model = None

for num_centers in param_grid['num_centers']:
    for gamma in param_grid['gamma']:
        for batch_size in param_grid['batch_size']:
            for epochs in param_grid['epochs']:
                print(f"Training with num_centers={num_centers}, gamma={gamma}, batch_size={batch_size}, epochs={epochs}")

                # Model oluşturma
                model = Sequential([
                    Input(shape=(X_train_scaled.shape[1],)),
                    RBFLayer(num_centers=num_centers, gamma=gamma),
                    Dense(43, activation='softmax')  # 43 sınıf için
                ])

                # Model derleme ve eğitim
                model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
                model.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, verbose=0)

                # Test performansını ölçme
                y_pred = np.argmax(model.predict(X_test_scaled), axis=1)
                accuracy = accuracy_score(y_test, y_pred)
                print(f"Validation Accuracy: {accuracy}")

                # En iyi modeli kaydetme
                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_params = {'num_centers': num_centers, 'gamma': gamma, 'batch_size': batch_size, 'epochs': epochs}
                    best_model = model

print(f"Best Hyperparameters: {best_params}")
print(f"Best Validation Accuracy: {best_accuracy}")

# En iyi modelin performans değerlendirmesi
y_pred_best = np.argmax(best_model.predict(X_test_scaled), axis=1)
print("Best RBF Model Accuracy:", accuracy_score(y_test, y_pred_best))
print("Classification Report:\n", classification_report(y_test, y_pred_best))

```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=True, yticklabels=True)
    plt.title(f"{title} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# MLP Modeli için confusion matrix
plot_confusion_matrix(y_test, best_mlp.predict(X_test_scaled), "MLP Model")

# RBF Modeli için confusion matrix
plot_confusion_matrix(y_test, y_pred, "RBF Model")
```
